```python
import pandas as pd
import json
```

```python
train0_df = pd.read_json('./0_train.json', lines=True)
train1_df = pd.read_json('./1_train.json', lines=True)

train_df = pd.concat([train0_df, train1_df], axis=0)
train_df = train_df.loc[:, ['answer', 'option_list', 'statement']]
train_df
```

```python
test_instances = []
for index, row in train_df.iterrows():
    instance_dict = {}
    answers, options, statement = row['answer'], row['option_list'], row['statement']
    opt_a, opt_b, opt_c, opt_d = options.values()
    instruction = f"{statement} (A) {opt_a} (B) {opt_b} (C) {opt_c} (D) {opt_d}."
    output = ''
    for answer in answers:
        if answer in ['A', 'B', 'C', 'D']:
            output += f"({answer}) {options[answer]}"
    output = f"{output}"
    test_instances.append({'instruction': instruction, 'input': '', 'output': output})
test_instances[:10]
```

```python
with open('qa_thunlp.json', 'w') as f:
    json.dump(test_instances, f, ensure_ascii=False)
```
